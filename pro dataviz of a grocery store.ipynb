## 18) Adding Columns:

- **Advantage:** It might help you to quickly assess an individual's overall financial position by combining both immediate earning power (salary) and long-term financial health (net worth).

* **Usefulness:** If you're dealing with large datasets and want a quick comparison, this combined average could highlight individuals who are relatively balanced in both earning and wealth accumulation.

###

```python
list_ = []

for i in range(0, 1000):

    row = df.iloc[i]
    a = row['Salary']
    b = row['Net Worth']
    c = (a + b) / 2
    list_.append(c)
#list_

df.insert(17, column = 'Mean ((Salary+Networth)/2)', value = list_)
df
```

```python
print('Find a row having 1 or 2 certain characters in a certain column!\n ')
column = str(input('Enter column: '))
value1 = input('Enter value1: ')
value2 = input('Enter value2: ')
value3 = input('Enter value3: ')

value_datatype = str(input('Enter Value1, Value2, and Value3 Data Type: '))

# Convert input values to specified datatype
if value_datatype == 'int':
    
    value1 = int(value1)
    value2 = int(value2)
    value3 = int(value3)
    
elif value_datatype == 'float':
    
    value1 = float(value1)
    value2 = float(value2)
    value3 = float(value3)
    
elif value_datatype == 'str':
    
    value1 = str(value1)
    value2 = str(value2)
    value3 = str(value3)
    
else:
    print('Invalid Input!')

# Combine values into a regex pattern
pattern = f"({value1}|{value2}|{value3})"

# Filter DataFrame
filtered_df = df[df[column].str.contains(pattern, na=False)]

# Display the result
print("Filtered DataFrame:")
filtered_df
```
